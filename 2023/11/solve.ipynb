```python
def remove_newline(values:list[str]):
    return [line.strip('\n') for line in values]
```

```python
from copy import deepcopy

def print_values(values:list):
    res = ''
    for row in values:
        res = res + ''.join(row)+'\n'
    print(res)

def parse_input(values:list[str]):
    len_x = len(values[0])
    len_y = len(values)

    values = deepcopy(values)

    for y in range(len_y):
        values[y] = list(values[y])

    print_values(values)

    # expand board
    # check columns that have to be expanded

    col_to_expand = list()
    for x in range(len_x):
        column = set()
        for y in range(len_y):
            column.add(values[y][x])
        if len(column) == 1:
            col_to_expand.append(x)
    
    row_to_expand = list()
    for y in range(len_y):
        row = set()
        for x in range(len_x):
            row.add(values[y][x])
        if len(row) == 1:
            row_to_expand.append(y)

    empty_row = ['.'] * len_x

    for idx, row in enumerate(row_to_expand):
        values.insert(row+idx, empty_row)

    for y in range(len(values)):
        for idx, col in enumerate(col_to_expand):
            values[y].insert(col+idx, '.')

    for y in range(len(values)):
        values[y] = values[y][0:len(values[0])] 

    len_x = len(values[0])
    len_y = len(values)

    # get coords of #
    planets_coords = set()
    for y in range(len_y):
        for x in range(len_x):
            if values[y][x] == '#':
                planets_coords.add((x,y))

    print_values(values)
    print(planets_coords)

    return values, planets_coords
```

```python
from itertools import combinations

def part_1(values:list[str]):
    values, planet_coords = parse_input(values)
    sum = 0
    for first, second in combinations(planet_coords, 2):
        diff_x = abs(first[0] - second[0])
        diff_y = abs(first[1] - second[1])
        sum += diff_x + diff_y
    return sum

part_1(test_values)
```
